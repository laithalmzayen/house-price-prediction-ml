```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.svm import SVR
```

```python
df=pd.read_csv('desktop/UserCarData.csv')
```

```python
df
```

```python
df=df.drop(columns="owner")
```

```python
z=pd.get_dummies(df,columns=["Region","State or Province","City","fuel","seller_type","transmission","torque","sold"])
```

```python
z
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
x=z.drop(columns="name")
y=z["name"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
clf=DecisionTreeClassifier(max_depth=10)
```

```python
clf.fit(xtrain,ytrain)
```

```python
yhat=clf.predict(xtest)
```

```python
clf.score(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
import matplotlib.pyplot as plt
df.plot('name', 'selling_price', kind='line')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns 
from sklearn.metrics import *
confusion_mc = confusion_matrix(ytest,yhat)
df_cm = pd.DataFrame(confusion_mc, 
                     index = [i for i in range(0,27)], columns = [i for i in range(0,27)])

plt.figure(figsize=(10,10))
sns.heatmap(df_cm, annot=True)
plt.title('DecisionTreeClassifier\nAccuracy:{0:.3f}'.format(accuracy_score(ytest, 
                                                                       yhat)))
plt.ylabel('True label')
plt.xlabel('Predicted label')
```
